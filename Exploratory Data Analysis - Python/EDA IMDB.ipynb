# Exploratory Data Analysis of IMDB Movies Dataset using Python 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the IMDB movies data

movies = pd.read_csv("Movie Assignment Data.csv")
pd.set_option('display.max_columns',70)
movies.head()
```

```python
#understanding the structure of the data 

movies.shape
```

```python
movies.info()
```

```python
#summary statistics of numerical columns

movies.describe(include="all")
```

```python
#converting budget and gross columns from $ to million $ for better readability 
movies['budget'] = movies['budget'].floordiv(1000000)
movies['Gross'] = movies['Gross'].floordiv(1000000)
```

```python
movies.head()
```

### Profit 
- create a new column Profit = Gross - budget 
- sort the data according to profit and finding the top 10 profitable movies 

```python
movies.insert(4, column = 'Profit', value = movies['Gross'] - movies['budget'])
```

```python
movies.head()
```

```python
movies.sort_values(by=['Profit'], ascending=False, inplace = True, ignore_index = True)
top10_movies = movies.loc[0:9,'Title']
top10_movies
```

```python
plt.figure(figsize=[8,6])
h = sns.jointplot(x='budget',y='Profit', data = movies)
h.set_axis_labels('Budget','Profit',fontsize=12)
plt.show()
```

#### Inference
- Most of the movies are profitable, there is a slight positive trend between 'budget' and 'profit'
- There are some big budget movies with negative profits. 
- between a budget of 0 - 100 million, there are a lot of profitable movies with upto 300 million profit.

## Worst Performing Movies (Negative Profit)

```python
negative_profit = movies[movies['Profit']<0]
negative_profit.reset_index(drop = True)
#Worst performing 5 movies
negative_profit.sort_values(by='Profit')[0:5]
```

### Movie Ratings Analysis
- MetaCritic (Critics opinion) rating and IMDB rating (public opinion)
- Average overall rating of a movie will be the average of both the ratings

```python
# MetaCritic rating is out of 100, so to make it into the same scale as IMDB rating
movies['MetaCritic'] = movies['MetaCritic']/10
```

```python
movies.insert(18,column='Avg Rating', value = (movies['MetaCritic']+movies['IMDb_rating'])/2)
```

```python
movies.head()
```

```python
#sorting movies by their avg rating (descending)
movies.sort_values(by='Avg Rating', ascending=False, inplace=True)
movies.head()
```

### Most Popular Actor Trios
- popoularity based on facebook likes of lead actors in the movie

```python
movies.insert(11,column='Popularity', value=movies.loc[:,['actor_1_facebook_likes','actor_2_facebook_likes'
                                                         ,'actor_3_facebook_likes']].sum(axis=1))
```

```python
#top 5 
most_popular_actors = movies.sort_values(by='Popularity',ascending=False, ignore_index=True).loc[0:4,:]
most_popular_actors
```

```python
most_popular_actor_trios = movies.sort_values(by='Popularity',ascending=False, ignore_index=True).loc[0:4,['actor_1_name','actor_2_name','actor_3_name']].values.tolist()
most_popular_actor_trios    
```

```python
#none of the three actors‚Äô Facebook likes should be less than half of the other two

individual_popular_trio = movies[~((((movies['actor_1_facebook_likes'] < movies['actor_2_facebook_likes']/2)|
                                     (movies['actor_1_facebook_likes'] < movies['actor_3_facebook_likes']/2))==True)
                                   |(((movies['actor_2_facebook_likes'] < movies['actor_1_facebook_likes']/2)|
                                     (movies['actor_2_facebook_likes'] < movies['actor_3_facebook_likes']/2))==True)
                                    |(((movies['actor_3_facebook_likes'] < movies['actor_1_facebook_likes']/2)|
                                     (movies['actor_3_facebook_likes'] < movies['actor_2_facebook_likes']/2))==True))]
```

```python
#common in the two dfs

common_actors  = most_popular_actors[most_popular_actors['Title'].isin(individual_popular_trio['Title'])]
common_actors.loc[:,['actor_1_name','actor_2_name','actor_3_name']].values.tolist()
```

```python
top_popular = individual_popular_trio.sort_values(by='Popularity',ascending=False)[0:5].reset_index(drop=True)
top_popular.loc[:,['actor_1_name','actor_2_name','actor_3_name']].values.tolist()
```

## Movie Genre Analysis

```python
#creating dataframe by genre
columns = []

for i in movies.columns:
    if i.startswith('CV')|i.startswith('V')|i.startswith("ge"):
        columns.append(i)
        
df_genre = movies.loc[:,columns]

df_genre['cnt'] = 1
#grouping movies by individual genre
df_g1 = df_genre.groupby('genre_1')
df_g2 = df_genre.groupby('genre_2')
df_g3 = df_genre.groupby('genre_3')

#finding the sum for genre1
df_g1 = pd.DataFrame(df_g1.sum())
df_g2 = pd.DataFrame(df_g2.sum())
df_g3 = pd.DataFrame(df_g3.sum())
```

```python
add_df1_df2 = df_g1.add(df_g2,fill_value=0)
df_add=add_df1_df2.add(df_g3,fill_value=0)

genre_top10 = df_add[df_add['cnt'] >=10]
genre_top10.head()
```

```python
#extracting count column
count_column= genre_top10['cnt']
#taking the mean of every column
genre_top10=genre_top10.div(genre_top10['cnt'],axis='index')

genre_top10=genre_top10.apply(lambda x:round(x,2),axis=0)
genre_top10.drop(columns='cnt',inplace=True)
```

```python
#convert CVotes into int

col_cvotes=[]
for i in genre_top10.columns:
    if i.startswith('CVotes'):
        col_cvotes.append(i)
        
genre_top10[col_cvotes]=genre_top10[col_cvotes].astype('int')
genre_top10.head()
```

```python
cnt = count_column.tolist()
genre_top10['cnt'] = cnt
```

```python
plt.figure(figsize=[10,6])
sns.barplot(x=genre_top10.index, y = genre_top10.cnt) 
plt.title('Genres with count',fontsize = 20)
plt.xlabel('Genre',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

## Top 1000 Voters in IMDB  vs Genre

```python
genre_top10_cvotes = genre_top10.sort_values(by='CVotes1000',ascending = False)
```

```python
plt.figure(figsize=[10,6])
sns.barplot(x=genre_top10_cvotes.index, y = genre_top10_cvotes.CVotes1000) 
plt.title('Top 1000 voters vs Genre',fontsize = 20)
plt.xlabel('Movie Genre',fontsize=15)
plt.ylabel('CV1000 Votes',fontsize=15)
plt.show()
```

### Inferences
- Romance genre was voted the least by top 1000 voters
- Sci-Fi is the most popular
- Action, Thriller and Adventure are similar, as they are genres that goes hand in hand
